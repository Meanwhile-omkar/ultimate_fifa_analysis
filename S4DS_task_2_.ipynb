```python
import datetime

def format_season(season):
  if isinstance(season, datetime.datetime):
    return season.strftime('%d/%m')  # Format as day/month
  else:
    return season  # Return original value if not datetime

data['Season'] = data['Season'].astype(str).apply(format_season)

incorrect_seasons = {'2024-12-11 00:00:00': '11/12', '2024-12-13 00:00:00': '12/13'}  # Add your incorrect and correct string values here
data['Season'] = data['Season'].replace(incorrect_seasons)
```

```python
data['Season'] = data['Season'].str.replace('O', '0')
```

# Data is Completed Cleaning and now its time for **Visualisation**

```python
plt.figure(figsize=(15,6))
goals_per_season = data.groupby('Season')['Type'].count()
goals_per_season.plot(kind='line', marker='o')
plt.xlabel('Season')
plt.ylabel('Number of Goals')
plt.title('Player Growth - Goals per Season')
plt.xticks(ticks=range(len(goals_per_season.index)), labels=goals_per_season.index, rotation=45)
plt.show()
```

```python
goals_per_club = data.groupby('Club')['Type'].count()
goals_per_club.plot(kind='bar')
plt.xlabel('Club')
plt.ylabel('Number of Goals')
plt.title('Goals Scored per Club')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python

goals_by_season_assist = data.groupby(['Season', 'Goal_assist'])['Type'].count().unstack(fill_value=0)

# Plot stacked bar chart
goals_by_season_assist.plot(kind='bar', stacked=True)
plt.xlabel('Season')
plt.ylabel('Number of Goals')
plt.title('Assisted vs. Non-Assisted Goals per Season')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Goal Assist')
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

goal_type_counts = data['Type'].value_counts()

# Create a bar chart
fig = px.bar(
    x=goal_type_counts.index,
    y=goal_type_counts.values,
    labels={'x': 'Goal Type', 'y': 'Number of Goals'},
    title='Distribution of Goal Types'
)

fig.show()
```

```python
goals_by_venue = data.groupby('Venue')['Type'].count()
goals_by_venue.plot(kind='pie', autopct='%1.1f%%')
plt.title('Goals Scored: Home vs. Abroad')
plt.ylabel('')  # Remove the y-axis label
plt.show()
```

```python
import plotly.express as px

# Group data, count goals, and sort by total goals in descending order
goals_by_position_type = data.groupby(['Playing_Position', 'Type'])['Type'].count().unstack(fill_value=0)
goals_by_position_type['Total'] = goals_by_position_type.sum(axis=1)
goals_by_position_type = goals_by_position_type.sort_values('Total', ascending=False)
goals_by_position_type = goals_by_position_type.drop('Total', axis=1)

# Create stacked bar chart with Plotly
fig = px.bar(
    goals_by_position_type,
    x=goals_by_position_type.index,
    y=goals_by_position_type.columns,
    title='Goal Types by Playing Position (Descending Order)',
    labels={'x': 'Playing Position', 'y': 'Number of Goals', 'variable': 'Goal Type'},
)

fig.update_layout(barmode='stack')
fig.show()
```

```python
# Group data by season and competition, count goals
goals_by_league_season = data.groupby(['Season', 'Competition'])['Type'].count().reset_index(name='Goals')

# Create grouped bar chart
fig = px.bar(
    goals_by_league_season,
    x='Season',
    y='Goals',
    color='Competition',
    barmode='group',
    title='Player Progress Across Leagues',
    labels={'Goals': 'Number of Goals'}
)

fig.show()
```
